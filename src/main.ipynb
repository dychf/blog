```python
import os
import sys
import yaml
from invest.stock import Stock
import pandas as pd
import logging

os.chdir(sys.path[0])

logger = logging.getLogger(__name__)

with open("config.yaml", "r", encoding="utf-8") as file:
    config = yaml.safe_load(file)

columns = ["name", "price", "V", "sell", "buy", "pe_avg", "period"]
data = []
for s_info in config["stock_codes"]:
    try:
        stock_info = Stock(s_info["code"]).valuation1()
        data.append(
            [
                stock_info.name,
                stock_info.market_price,
                stock_info.valuation_price,
                stock_info.ideal_sell,
                stock_info.ideal_buy,
                stock_info.industry_mean_pe,
                stock_info.period,
            ]
        )

    except Exception as e:
        logger.error(f"An error occurred: {e}")
        continue
df = pd.DataFrame(data, columns=columns)
df
```
